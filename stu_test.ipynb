```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


import src.STU_2hdms as stu
import matplotlib as mpl
```

```python
class inpind:
    ix = 2
    iy = 9
    mH = 800
    mh = 125
    mhs = 800
    mA = 800
    mas = 800
    mhp = 800
    tb = 1
    ba = 0
    aHS = 0
    ahs = 0
    aAS = 0


def debug(npda):
    lst = []
    for li in npda:
        lst.append(list(li))
    return(lst)


def stu_calc(x, y ):
    inp = [
        inpind.mH, 
        inpind.mh, 
        inpind.mhs,
        inpind.mA,
        inpind.mas,
        inpind.mhp,
        inpind.tb,
        0.0,
        inpind.aHS,
        inpind.ahs,
        inpind.aAS,
        ]

    if inpind.ix == 7:
        inp[7] = np.arctan(inpind.tb)- (np.pi/2 - x)
    else : 
        inp[7] = np.arctan(inpind.tb)- (np.pi/2 - inpind.ba)
        inp[inpind.ix] = x
    if inpind.iy == 7:
        inp[7] = np.arctan(inpind.tb)- (np.pi/2 - y)
    else:
        inp[7] = np.arctan(inpind.tb)- (np.pi/2 - inpind.ba)
        inp[inpind.iy] = y

    # print(inp)
    stu.STUchk(*inp)
    return stu.STU.chi2

def cont(lx,ly):
    XX, YY = np.meshgrid(lx, ly)
    ct = np.frompyfunc(stu_calc, 2, 1)
    return XX, YY, debug(ct(XX,YY))
```

```python




fig, axs = plt.subplots(4,4, figsize=(10,8))

k = 0
l = 0



for nx in [0,2,3,4]:
    for ny in [7,8,9,10]:
        inpind.ix = nx
        inpind.iy = ny
        ztup = cont(np.linspace(100, 2000), np.linspace(-np.pi/2, np.pi/2))
        axs[k][l].contourf(*ztup, levels=[0, 7.81],alpha=0.1, cmap=mpl.colors.ListedColormap(['white','blue']))
        cp = axs[k][l].contour(*ztup, levels=[7.81], colors=['tab:blue'])
        k = k+1
    l = l+1
    k = 0

axs[0][0].set_ylabel(r'$c_{\beta-\alpha}$')
axs[1][0].set_ylabel(r'$\alpha_{HS}$')
axs[2][0].set_ylabel(r'$\alpha_{hS}$')
axs[3][0].set_ylabel(r'$\alpha_{AS}$')
axs[3][0].set_xlabel(r'$m_H$')
axs[3][1].set_xlabel(r'$m_{h_S}$')
axs[3][2].set_xlabel(r'$m_A$')
axs[3][3].set_xlabel(r'$m_{A_S}$')
```
